```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import col
spark = SparkSession.builder.getOrCreate()
```

```python
def snake_case_cols(df):
    for c in df.columns:
        df=df.withColumnRenamed(
            c, c.lower().replace(" ","_").replace("-","_")
        )
    return df
```

```python
players_silver = spark.read.parquet("/home/gnana/football-lakehouse/silver/players_silver")
```

```python
players_silver = snake_case_cols(players_silver)
```
