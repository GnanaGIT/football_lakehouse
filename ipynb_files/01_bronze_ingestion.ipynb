```python
from pyspark.sql import SparkSession
import os
import pandas as pd
spark = SparkSession.builder.appName("football-bronze").getOrCreate()
```

```python
raw_path = "/home/gnana/football-lakehouse/raw_csv"
bronze_path = "/home/gnana/football-lakehouse/bronze"
```

```python
csv_files = [f for f in os.listdir(raw_path) if f.endswith('csv')]
csv_files
```

```python
for file in csv_files:
    table = file.replace(".csv", "")
    data = spark.read.option("header", True).option('inferSchema', True).csv(f"{raw_path}/{file}")
    data.write.mode("overwrite").parquet(f"{bronze_path}/{table}")
```
